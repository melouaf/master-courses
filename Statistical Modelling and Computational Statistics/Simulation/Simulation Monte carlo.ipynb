```
# Import basic packages

import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```
def Mean(x):                              
  return np.mean(np.array(x))

def Var(x):                        
  return np.var(x)
```

# Première fonction

```
def g(x):
    return np.sinh((1/3)+np.sqrt(x))/(np.sinh((1/3)+1))  ##on normalise la fonction de sorte que ses valeurs soient dans l'intervalle [0,1]
                                                         #sinon l'estimateur de la fonction U n'aura pas de sens.

def V(x):
  return g(x)

def U(x,y):
    if y<=g(x):
        return 1
    return 0

def W(x):
  return (g(x)+g(1-x))/2
```

```

N_values=np.linspace(200,10000,300,endpoint=True) 

u_mean =[]                                                       ## Estimation par la fonction U
u_var=[]  
v_mean =[]                                                       ##Estimation par la fonction V
v_var =[]
w_mean =[]                                                       ## Estimation par la fonction W
w_var =[]

for i in range(len(N_values)):
  u_values=[]
  v_values=[]
  w_values=[]
  for j in  range(int(N_values[i])):

    u_values.append(U(np.random.uniform(0, 1), np.random.uniform(0,1)))  ## Estimation par la fonction U
    v_values.append(V(np.random.uniform(0,1)))                           ##Estimation par la fonction V
    w_values.append(W(np.random.uniform(0,1)))                           ##Estimation par la fonction W

  u_mean.append(Mean(u_values)*(np.sinh((1/3)+1)))
  u_var.append((Var(u_values)/int(N_values[i]))*((np.sinh((1/3)+1))**2))

  v_mean.append(Mean(v_values)*(np.sinh((1/3)+1)))                                ## dénormalisation 
  v_var.append((Var(v_values)/int(N_values[i]))*((np.sinh((1/3)+1))**2))          ## dénormalisation 

  w_mean.append(Mean(w_values)*(np.sinh((1/3)+1)))
  w_var.append((Var(w_values)/int(N_values[i]))*((np.sinh((1/3)+1))**2))
```

### Diagrames

```
## Comparaison des diagrames 

#Moyenne

plt.plot(N_values, v_mean, label="La moyenne pour l'estimateur de V")
plt.plot(N_values, w_mean, label="La moyenne pour l'estimateur de W")
plt.plot(N_values, u_mean, label="La moyenne pour l'estimateur de U")
plt.xlabel("Nombre d'échantillons")
plt.ylabel("Moyenne")

plt.title("Comparaison : la Moyenne des estimateurs.")
plt.legend()
plt.show()


## Variance

plt.plot(N_values, v_var, label="La variance pour l'estimateur de V")
plt.plot(N_values, w_var, label="La varaiance pour l'estimateur de W")
plt.plot(N_values, u_var, label="La variance pour l'estimateur de U")
plt.xlabel("Nombre d'échantillons")
plt.ylabel("Moyenne")

plt.title("Comparaison : la variance des estimateurs.")
plt.legend()
plt.show()
```

# Deuxième fonction

```
def h(x):
    return (np.log(1+3*x**(3.2))**(1/4))/(np.log(1+3*1**(3.2))**(1/4))     #On prend la fonction opposée renormaliséeafin que ses valeurs soient dans [0,1]
                                                                           #sinon l'estimateur de la fonction U n'aura pas de sens.

def V1(x):
  return h(x)

def U1(x,y):
    if y<=h(x):
        return 1
    return 0

def W1(x):
  return (h(x)+h(1-x))/2
```
